```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dsf_data = pd.read_csv("dsf_2004_2022.csv")
```

```python
dsf_data['date'] = pd.to_datetime(dsf_data['date'])
```

# 2.1

```python
dsf_data['RET'] = pd.to_numeric(dsf_data['RET'], errors='coerce')
dsf_data = dsf_data.dropna(subset=['RET'])

dsf_data['PRC'] = dsf_data['PRC'].abs()

dsf_data['Bid-Ask_Spread'] = dsf_data['ASK'] - dsf_data['BID']

dsf_data['Range of Stocks-high/low'] = dsf_data['ASKHI'] - dsf_data['BIDLO']
dsf_data['Range of Stocks-open/close'] = dsf_data['PRC'] - dsf_data['OPENPRC']
```

```python
selected_columns = ['PERMNO','date','EXCHCD','SHRCD','BIDLO','ASKHI','PRC','VOL','RET','BID','ASK','SHROUT','OPENPRC','vwretd','Bid-Ask_Spread','Range of Stocks-high/low','Range of Stocks-open/close'] 
dsf_data_1 = dsf_data[selected_columns].copy()
```

```python
def calculate_stats(df_original, columns):
    # Create an empty DataFrame to store results
    res_df = pd.DataFrame()

    for col in columns:
        # Calculate basic statistics for the column
        col_stats = df_original[col].describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99])

        # Calculate skewness and kurtosis
        skewness = df_original[col].skew()
        kurtosis = df_original[col].kurtosis()

        # Create a DataFrame with column-specific statistics
        col_stats = pd.DataFrame(col_stats).T

        # Add column name, skewness, and kurtosis to the DataFrame
        col_stats['skewness'] = skewness
        col_stats['kurtosis'] = kurtosis

        # Append the column-specific statistics to the results DataFrame
        res_df = pd.concat([res_df, col_stats])

    return res_df
```

```python
column = ['RET', 'PRC', 'Bid-Ask_Spread','Range of Stocks-high/low','Range of Stocks-open/close']
```

```python
def plot_data(df_original, date_col='date', columns=None, freq='M'):
    # Create a copy of the DataFrame to avoid modifying the original
    df_copy = df_original.copy()
    
    # Convert the date column to datetime
    df_copy[date_col] = pd.to_datetime(df_copy[date_col])
    df_copy.set_index(date_col, inplace=True)
    
    if columns is None:
        columns = [col for col in df_copy.columns if col != date_col]
    
    resampled = df_copy[columns].resample(freq).mean()

    # Define custom line colors
    custom_colors = ['red', 'blue', 'green']
    
    ax = resampled.plot(figsize=(16, 8), grid=True, color=custom_colors)

    # Add a legend outside the graph
    ax.legend(columns, loc='center left', bbox_to_anchor=(1, 0.5))

    plt.xlabel(date_col)
    plt.ylabel('Value of Mean')
    plt.title(f"Resampled Plot of {', '.join(columns)} vs. {date_col}")
    plt.tight_layout()
    plt.show()
```

```python
plot_data(dsf_data_1, date_col='date', columns=['Range of Stocks-high/low','Range of Stocks-open/close', 'RET'], freq='M')
```

```python


# Make a copy of the original dataset to avoid modifying it
dsf_data_copy = dsf_data_1.copy()

# Assuming you have a DataFrame named filtered_dsf with columns: 'Date', 'PERMNO', and 'PRC'

# Convert the 'Date' column to datetime
dsf_data_copy['date'] = pd.to_datetime(dsf_data_copy['date'])

# Initialize an empty DataFrame to store the returns
returns_df = pd.DataFrame(columns=['PERMNO', 'Start Date', 'End Date', 'Start Price', 'End Price', 'Return'])

# Group data by PERMNO
grouped = dsf_data_copy.groupby('PERMNO')

# Function to calculate returns for each group
def calculate_returns(group):
    start_date = group['date'].min()
    end_date = start_date + pd.DateOffset(years=3)
    
    # Filter data for the first three years
    group = group[group['date'] <= end_date]
    
    if not group.empty:
        beginning_prc = group.loc[group['date'] == start_date, 'PRC'].values[0]
        end_prc = group.loc[group['date'] == group['date'].max(), 'PRC'].values[0]
        stock_return = (end_prc - beginning_prc) / beginning_prc
        return pd.Series({
            'PERMNO': group['PERMNO'].iloc[0],
            'Start Date': start_date,
            'End Date': group['date'].max(),
            'Start Price': beginning_prc,
            'End Price': end_prc,
            'Return': stock_return
        })

# Apply the function to each group and append results to returns_df
returns_df = grouped.apply(calculate_returns).reset_index(drop=True)

# Display the returns DataFrame
returns_df.head()
```

# 2.2

```python
dsf_data_copy_2 = dsf_data_1.copy()
dsf_data_copy_2['date'] = pd.to_datetime(dsf_data_copy_2['date'])

ipo_dates = dsf_data_copy_2.groupby('PERMNO')['date'].min().reset_index()
ipo_dates.columns = ['PERMNO', 'IPO_Date']
date_range = pd.date_range(start='2004-01-01', end='2022-12-31', freq='M')
month_having_ipos = []

for month in date_range:
    ipos_in_month = ipo_dates[ipo_dates['IPO_Date'].dt.month == month.month]
    if not ipos_in_month.empty:
        month_having_ipos.append({'date': month, 'IPO Count': len(ipos_in_month)})

ipo_months = pd.DataFrame(month_having_ipos)

ipo_months['date'] = pd.to_datetime(ipo_months['date'])

dsf_data_copy_2.set_index('date', inplace=True)

vwretd_monthly = dsf_data_copy_2['vwretd'].resample('M').mean()
vwretd_monthly = vwretd_monthly.reset_index()

ipo = ipo_months.merge(vwretd_monthly, left_on='date', right_on='date', how='left')
```

```python
ipo.head(10)
```

```python
fig, ax1 = plt.subplots(figsize=(16, 8))
sns.set_style("whitegrid")


sns.lineplot(data=ipo, x='date', y='IPO Count', ax=ax1, label='IPO Count',color='b')
ax1.set_xlabel('Date')

ax1.tick_params(axis='y', labelcolor='r')

ax2 = ax1.twinx()

sns.lineplot(data=ipo, x='date', y='vwretd', ax=ax2, label='VWRETD',color='g')

ax2.tick_params(axis='y', labelcolor='y')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels+labels2, loc='lower right')

plt.title('IPO Count and VWRETD Over Time')
plt.show()
```

# 2.3

```python
dsf_data_copy_3 = dsf_data_1.copy()

dsf_data_copy_3['date'] = pd.to_datetime(dsf_data_copy_3['date'])
dsf_data_copy_3.set_index('date', inplace=True)

data_by_month = dsf_data_copy_3.resample('M').agg({'PRC': 'mean', 'SHROUT': 'mean', 'vwretd': 'mean'})

data_by_month['Average Market Capitalization'] = data_by_month['PRC'].abs() * data_by_month['SHROUT']

# Reset the index to make 'Date' a column
data_by_month.reset_index(inplace=True)
```

```python
data_by_month.head()
```

```python
fig, ax1 = plt.subplots(figsize=(16,8))
sns.lineplot(data=data_by_month, x='date', y='Average Market Capitalization', ax=ax1, label='Average Market Capitalization', color='pink')
ax2 = ax1.twinx()
sns.lineplot(data=data_by_month, x='date', y='vwretd', ax=ax2, label='VWRETD', color='green')

ax1.set_ylabel('Average Market Capitalization')
ax2.set_ylabel('VWRETD')
ax1.set_xlabel('date')
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1+labels2, loc='lower right')
ax2.legend()

plt.title('Average Market Capitalization and VWRETD Over Time')

plt.show()
```

```python
dsf_data_copy_4 = dsf_data_1.copy()
res_df = pd.DataFrame(columns=['PERMNO', 'RET'])
dsf_data_copy_4['date'] = pd.to_datetime(dsf_data_copy_4['date'])

# Loop over unique PERMNOs to process each stock's returns
unique_val = dsf_data_copy_4['PERMNO'].unique()

for permno in unique_val:
    stock_data = dsf_data_copy_4[dsf_data_copy_4['PERMNO'] == permno]
    start_date = stock_data['date'].min()

    # Calculate the end_date as 3 years from the start_date
    end_date = start_date

    # Find the nearest date to end_date that is less than or equal to end_date
    nearest_date = stock_data[stock_data['date'] <= end_date]['date'].max()

    # Check if nearest_date is within 3 years of start_date
    if nearest_date is not pd.NaT:
        beginning_prc = stock_data[stock_data['date'] == start_date]['OPENPRC'].values[0]
        ending_prc = stock_data[stock_data['date'] == nearest_date]['PRC'].values[0]
        stock_return = (ending_prc - beginning_prc) / beginning_prc
        res_df = pd.concat([res_df, pd.DataFrame({'PERMNO': [permno], 'Date': [start_date], 'End Date': [nearest_date], 'Start Price': [beginning_prc], 'End Price': [ending_prc], 'Mean - IPO Returns': [stock_return]})], ignore_index=True)

# Calculate the sum of returns for each unique "Date"
mean_ipo = res_df.groupby(["Date"])["Mean - IPO Returns"].mean().reset_index()

mean_ipo.head()
```

```python
fig, ax1 = plt.subplots(figsize=(16,8))
sns.lineplot(data=mean_ipo, x='Date', y='Mean - IPO Returns', ax=ax1, label='Mean - IPO Returns', color='pink')
ax2 = ax1.twinx()
sns.lineplot(data=data_by_month, x='date', y='vwretd', ax=ax2, label='VWRETD', color='green')

ax1.set_ylabel('Mean - IPO Returns')
ax2.set_ylabel('VWRETD')
ax1.set_xlabel('date')
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1+labels2, loc='lower right')
ax2.legend()

plt.title('Mean - IPO Returns and VWRETD Over Time')

plt.show()
```
