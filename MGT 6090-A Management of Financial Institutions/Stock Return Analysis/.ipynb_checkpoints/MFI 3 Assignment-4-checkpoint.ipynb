```python
def plot_resampled_data(df, date_col='Date', columns=None, freq ='M'):
    
    df[date_col] = pd.to_datetime(df[date_col])
    df.set_index(date_col, inplace=True)
    
    if columns is None:
        columns = [col for col in df.columns if col != date_col]
    
    resampled = df[columns].resample(freq).mean()

    ax = resampled.plot(figsize=(15, 7), grid=True)
    
    plt.xlabel(date_col)
    plt.ylabel('Mean Value')
    plt.title(f"Resampled Plot of {', '.join(columns)} vs. {date_col}")
    plt.tight_layout()
    plt.show()
```

```python
plot_resampled_data(filtered_msf,date_col='Date',columns= ["Inflated Market Cap", "Market Cap"], freq ='Y')
```

```python
dates = filtered_msf.groupby('Date').sum(numeric_only=True).reset_index()['Date']

RET_st = pd.DataFrame()
CPI_st = pd.DataFrame()
InflatedMarketCap_st = pd.DataFrame()



stats = {
    'N': np.nan,
    'Mean': np.nan,
    'Standard Deviation': np.nan,
    'Skewness': np.nan,
    'Kurtosis': np.nan,
    'Min': np.nan,
    'Max': np.nan,
    '1% Percentile': np.nan,
    '5% Percentile': np.nan,
    '25% Percentile': np.nan,
    '50% Percentile': np.nan,
    '75% Percentile': np.nan,
    '95% Percentile': np.nan,
    '99% Percentile': np.nan
}

for df in [RET_st, CPI_st, InflatedMarketCap_st]:
    df['Date'] = np.nan
    for x in stats:
        df[x] = np.nan
```

```python
def compute_statistics(column, date):
    # Basic statistics
    N = len(column)
    mean = column.mean()
    std = column.std()
    skewness = column.skew()
    kurtosis_val = column.kurtosis()
    
    # Percentiles
    min_val = column.min()
    max_val = column.max()
    p1 = column.quantile(0.01)
    p5 = column.quantile(0.05)
    p25 = column.quantile(0.25)
    p50 = column.quantile(0.50)
    p75 = column.quantile(0.75)
    p95 = column.quantile(0.95)
    p99 = column.quantile(0.99)
    return {
        "Date": date,
        'N': N,
        'Mean': mean,
        'Standard Deviation': std,
        'Skewness': skewness,
        'Kurtosis': kurtosis_val,
        'Min': min_val,
        'Max': max_val,
        '1% Percentile': p1,
        '5% Percentile': p5,
        '25% Percentile': p25,
        '50% Percentile': p50,
        '75% Percentile': p75,
        '95% Percentile': p95,
        '99% Percentile': p99
    }

# Create lists to collect statistics
RET_stats_list = []
CPI_stats_list = []
IMC_stats_list = []


for date in dates:
    subset = filtered_msf.copy()[filtered_msf['Date'] == date][['RET','CPI Index','Inflated Market Cap']]
    
    # Compute statistics for each subset and append to respective lists
    RET_stats_list.append(compute_statistics(subset['RET'], date))
    CPI_stats_list.append(compute_statistics(subset['CPI Index'], date))
    IMC_stats_list.append(compute_statistics(subset['Inflated Market Cap'], date))
    
    
    

# Create DataFrames from the lists of statistics
RET_stats = pd.DataFrame(RET_stats_list)
CPI_stats = pd.DataFrame(CPI_stats_list)
IMC_stats = pd.DataFrame(IMC_stats_list)

```

```python
def descriptive_stats(df_original, columns):
    # Create an empty list to store results
    results = []

    for col in columns:
        df = df_original.copy()

        stats = df[col].describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99])
        stats['skewness'] = df[col].skew()
        stats['kurtosis'] = df[col].kurtosis()
        
        # Add the column name as an attribute to the Series
        stats['column_name'] = col

        # Append the results for this column to the list
        results.append(stats)

    # Concatenate the results into a DataFrame
    results_df = pd.concat(results, axis=1).T
    
    return results_df
```

```python
def plot_resampled_data(df, date_col='Date', columns=None, freq='M'):
    # Create a copy of the DataFrame to avoid modifying the original
    df_copy = df.copy()
    
    # Convert the date column to datetime
    df_copy[date_col] = pd.to_datetime(df_copy[date_col])
    df_copy.set_index(date_col, inplace=True)
    
    if columns is None:
        columns = [col for col in df_copy.columns if col != date_col]
    
    resampled = df_copy[columns].resample(freq).mean()

    ax = resampled.plot(figsize=(15, 7), grid=True)
    
    plt.xlabel(date_col)
    plt.ylabel('Mean Value')
    plt.title(f"Resampled Plot of {', '.join(columns)} vs. {date_col}")
    plt.tight_layout()
    plt.show()

# Call the function with filtered_dsf

```

```python
plot_resampled_data(filtered_dsf, date_col='Date', columns=["Price Range", "Price Difference", "RET"], freq='M')
```

```python
filtered_dsf_copy1 = filtered_dsf.copy()

returns_df = pd.DataFrame(columns=['PERMNO', 'Return'])
filtered_dsf_copy1['Date'] = pd.to_datetime(filtered_dsf_copy1['Date'])

unique_permnos = filtered_dsf_copy1['PERMNO'].unique()
end_date_max = pd.to_datetime('2022-12-31')

for permno in unique_permnos:
    stock_data = filtered_dsf_copy1[filtered_dsf_copy1['PERMNO'] == permno]
    start_date = stock_data['Date'].min()
    end_date = start_date + pd.DateOffset(years=3)
    closest_date = stock_data[stock_data['Date'] <= end_date]['Date'].max()
    
    if closest_date is not pd.NaT and (closest_date - start_date).days <= 365 * 3:
        beginning_prc = stock_data[stock_data['Date'] == start_date]['PRC'].values[0]
        ending_prc = stock_data[stock_data['Date'] == closest_date]['PRC'].values[0]
        stock_return = (ending_prc - beginning_prc) / beginning_prc
        returns_df = pd.concat([returns_df, pd.DataFrame({'PERMNO': [permno], 'Start Date': [start_date], 'End Date': [closest_date], 'Start Price': [beginning_prc], 'End Price': [ending_prc], 'Return': [stock_return]})], ignore_index=True)
```

```python

filtered_dsf_copy = filtered_dsf.copy()
filtered_dsf_copy['Date'] = pd.to_datetime(filtered_dsf_copy['Date'])

ipo_dates = filtered_dsf_copy.groupby('PERMNO')['Date'].min().reset_index()
ipo_dates.columns = ['PERMNO', 'IPO_Date']

date_range = pd.date_range(start='2004-01-01', end='2022-12-31', freq='M')

months_with_ipos = []

for month in date_range:
    ipos_in_month = ipo_dates[ipo_dates['IPO_Date'].dt.month == month.month]
    if not ipos_in_month.empty:
        months_with_ipos.append({'Date': month, 'IPO Count': len(ipos_in_month)})

ipo_months_df = pd.DataFrame(months_with_ipos)


ipo_months_df['Date'] = pd.to_datetime(ipo_months_df['Date'])

filtered_dsf_copy.set_index('Date', inplace=True)


monthly_vwretd = filtered_dsf_copy['vwretd'].resample('M').mean()


monthly_vwretd = monthly_vwretd.reset_index()


ipo_df = ipo_months_df.merge(monthly_vwretd, left_on='Date', right_on='Date', how='left')





```

```python
ipo_df.head()
```

```python

fig, ax1 = plt.subplots(figsize=(12, 6))
sns.set_style("whitegrid")


sns.lineplot(data=ipo_df, x='Date', y='IPO Count', ax=ax1, label='IPO Count', color='teal')
ax1.set_xlabel('Date')

ax1.tick_params(axis='y', labelcolor='r')


ax2 = ax1.twinx()

sns.lineplot(data=ipo_df, x='Date', y='vwretd', ax=ax2, label='VWRETD', color='orange')

ax2.tick_params(axis='y', labelcolor='y')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend()

plt.title('IPO Count and VWRETD Over Time with Dual Y-Axes')
plt.show()
```

```python

filtered_dsf_copy4 = filtered_dsf_copy1.copy()


returns_df = pd.DataFrame(columns=['PERMNO', 'RET'])

filtered_dsf_copy4['Date'] = pd.to_datetime(filtered_dsf_copy4['Date'])

# Loop over unique PERMNOs to process each stock's returns
unique_permnos = filtered_dsf_copy4['PERMNO'].unique()

for permno in unique_permnos:
    stock_data = filtered_dsf_copy4[filtered_dsf_copy4['PERMNO'] == permno]
    start_date = stock_data['Date'].min()

    # Calculate the end_date as 3 years from the start_date
    end_date = start_date

    # Find the closest date to end_date that is less than or equal to end_date
    closest_date = stock_data[stock_data['Date'] <= end_date]['Date'].max()

    # Check if closest_date is within 3 years of start_date
    if closest_date is not pd.NaT:
        beginning_prc = stock_data[stock_data['Date'] == start_date]['OPENPRC'].values[0]
        ending_prc = stock_data[stock_data['Date'] == closest_date]['PRC'].values[0]
        stock_return = (ending_prc - beginning_prc) / beginning_prc
        returns_df = pd.concat([returns_df, pd.DataFrame({'PERMNO': [permno], 'Date': [start_date], 'End Date': [closest_date], 'Start Price': [beginning_prc], 'End Price': [ending_prc], 'Mean IPO Returns': [stock_return]})], ignore_index=True)

# Calculate the sum of returns for each unique "Start Date"
meanreturns_ipo = returns_df.groupby(["Date"])["Mean IPO Returns"].mean().reset_index()

meanreturns_ipo.head()
```
