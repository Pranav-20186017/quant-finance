```python
# Define a list of categorical columns that need placeholder filling
categorical_cols = ['SHRCD', 'EXCHCD', 'TICKER', 'COMNAM', 'HSICCD']

# Convert 'SHRCD' and 'EXCHCD' to categorical and add 'Unknown' as a category, then fill missing values with 'Unknown'
for col in categorical_cols:
    msf_data[col] = msf_data[col].astype('category')
    msf_data[col] = msf_data[col].cat.add_categories(['Unknown'])
    msf_data[col].fillna('Unknown', inplace=True)

# Define a list of continuous columns that havenâ€™t undergone linear interpolation
uninterpolated_cols = ['BIDLO', 'ASKHI', 'VOL', 'BID', 'ASK', 'SHROUT', 'CFACPR', 'CFACSHR'] 

# Interpolate or Forward-fill/Backward-fill for uninterpolated continuous variables
msf_data[uninterpolated_cols] = msf_data[uninterpolated_cols].interpolate(method='linear', limit_direction='both')

# Given the substantial amount of missing data in 'SPREAD', filling with the median value
msf_data['SPREAD'].fillna(msf_data['SPREAD'].median(), inplace=True)

# Convert the 'ALTPRCDT' to datetime, coerce errors to NaT
msf_data['ALTPRCDT'] = pd.to_datetime(msf_data['ALTPRCDT'], errors='coerce')

# Now find the minimum date ignoring NaT values
min_date = msf_data['ALTPRCDT'].min(skipna=True)

# Fill NaT values with the minimum date
msf_data['ALTPRCDT'].fillna(min_date, inplace=True)

# Define a list of columns that still have missing values after previous operations
remaining_na_cols = ['PRC', 'ALTPRC', 'vwretd']

# Interpolate or Forward-fill/Backward-fill for continuous variables with remaining missing values
msf_data[remaining_na_cols] = msf_data[remaining_na_cols].interpolate(method='linear', limit_direction='both')

# If interpolation does not fill all missing values, fill the remaining ones with median values of the respective columns
msf_data[remaining_na_cols] = msf_data[remaining_na_cols].apply(lambda x: x.fillna(x.median()))

# For 'RET' and 'RETX', consider filling with 0 as they are return columns and NaN might represent no return
msf_data[['RET', 'RETX']] = msf_data[['RET', 'RETX']].fillna(0)

# Check the DataFrame again for any remaining missing values and review if any further steps are needed.
print(msf_data.isna().sum())
```

## Number of Stocks and Market Capitalization over time

```python
# Filtering the merged_data
filtered_data = merged_data[(merged_data['date'] >= '1925-12') & (merged_data['date'] <= '2019-12')]

# Calculating the number of stocks each month
num_stocks = filtered_data.groupby('date')['PERMNO'].nunique().reset_index()
num_stocks.columns = ['date', 'num_stocks']

# Calculating the total market capitalization each month
total_mkt_cap = filtered_data.groupby('date')['SCALED_MKT_CAP'].sum().reset_index()
total_mkt_cap.columns = ['date', 'total_mkt_cap']

# Merging the two dataframes
plot_data = pd.merge(num_stocks, total_mkt_cap, on='date')

# Convert the 'date' column to datetime
plot_data['date'] = plot_data['date'].dt.to_timestamp()

# Plotting the data
fig, ax1 = plt.subplots(figsize=(14, 7))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Number of Stocks', color=color)
ax1.plot(plot_data['date'], plot_data['num_stocks'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Total Market Cap (Dec 2010 $)', color=color)
ax2.plot(plot_data['date'], plot_data['total_mkt_cap'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Number of Stocks and Market Capitalization over time (1925-2019)')
plt.show()
```

```python
# Define the bin edges and corresponding labels for the industries
bin_edges = [0, 999, 1499, 1799, 3999, 4999, 5199, 5999, 6799, 8999, 9999]
industry_labels = [
    'Agriculture, Forestry and Fishing',
    'Mining',
    'Construction',
    'Manufacturing',
    'Transportation and other Utilities',
    'Wholesale Trade',
    'Retail Trade',
    'Finance, Insurance and Real Estate',
    'Services',
    'Public Administration'
]

# Filter the data to the required date range
filtered_data = merged_data[(merged_data['date'] >= '1925-12') & (merged_data['date'] <= '2022-12')]

# Convert HSICCD to numeric (it might be object type due to 'Unknown' entries)
filtered_data['HSICCD'] = pd.to_numeric(filtered_data['HSICCD'], errors='coerce')

# Vectorized categorization of industries
filtered_data['Industry'] = pd.cut(filtered_data['HSICCD'], bins=bin_edges, labels=industry_labels, right=True, include_lowest=True)

# Add 'Unknown' as a category for the 'Industry' column before filling NaN values
filtered_data['Industry'] = filtered_data['Industry'].cat.add_categories('Unknown')
filtered_data['Industry'].fillna('Unknown', inplace=True)  # Now fill NaN values with 'Unknown'

# Group by date and industry and calculate the number of unique stocks and total market capitalization
grouped_data = filtered_data.groupby(['date', 'Industry']).agg({'PERMNO': 'nunique', 'SCALED_MKT_CAP': 'sum'}).reset_index()
grouped_data.columns = ['date', 'Industry', 'num_stocks', 'total_mkt_cap']

# Convert the 'date' column to datetime
grouped_data['date'] = grouped_data['date'].dt.to_timestamp()
```

```python
# Plotting
fig, axs = plt.subplots(2, 1, figsize=(14, 14), sharex=False)  # Set sharex to False

# Loop over the industry_labels list directly
for industry in industry_labels:
    industry_data = grouped_data[grouped_data['Industry'] == industry]
    axs[0].plot(industry_data['date'], industry_data['num_stocks'], label=industry)
    axs[1].plot(industry_data['date'], industry_data['total_mkt_cap'], label=industry)

# Configure the plot for Number of Stocks
axs[0].set_ylabel('Number of Stocks')
axs[0].set_title('Number of Stocks in Each Industry over time (1925-2022)')
axs[0].legend(loc='upper left')
axs[0].set_xlabel('Date')  # Set x-axis label for the first subplot

# Configure the plot for Market Capitalization
axs[1].set_ylabel('Total Market Cap (Dec 2010 $)')
axs[1].set_title('Market Capitalization in Each Industry over time (1925-2022)')
axs[1].legend(loc='upper left')

# Adjust the space between the plots
plt.subplots_adjust(hspace=0.5)  # Adjust this value as needed to avoid overlapping

plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
# Replace 'C' with NaN and convert to numeric
merged_data_final['RET'] = pd.to_numeric(merged_data_final['RET'], errors='coerce')
merged_data_final['RF'] = pd.to_numeric(merged_data_final['RF'], errors='coerce')

# Fill NaN with 0 or any specific value that suits your analysis
merged_data_final['RET'].fillna(0, inplace=True)
merged_data_final['RF'].fillna(0, inplace=True)

# Compute excess return and log of return
merged_data_final['excess_return'] = merged_data_final['RET'] - merged_data_final['RF']
merged_data_final['log_return'] = np.log1p(merged_data_final['RET'])
```

## Compute CRSP_MSF descriptive stats and plot monthly

```python
# Define the time periods
time_periods = {
    '1925-2022': ('1925-01-01', '2022-12-31'),
    '1963-2022': ('1963-01-01', '2022-12-31'),
}

# Initialize a dictionary to store the descriptive stats for each time period
descriptive_stats_dict = {}

for label, (start, end) in time_periods.items():
    # Filter the data for the specific time period
    filtered_data = merged_data_final[(merged_data_final['date'] >= start) & (merged_data_final['date'] <= end)]
    
    # Compute descriptive statistics with specific percentiles
    descriptive_stats = filtered_data['excess_return'].describe(
        percentiles=[.01, .05, .25, .5, .75, .95, .99]
    ).to_frame().rename(columns={'excess_return': label})
    
    # Compute skewness and kurtosis
    descriptive_stats.loc['skewness'] = [filtered_data['excess_return'].skew()]
    descriptive_stats.loc['kurtosis'] = [filtered_data['excess_return'].kurtosis()]
    
    # Store the descriptive stats in the dictionary
    descriptive_stats_dict[label] = descriptive_stats

# Convert the descriptive stats dictionary to a DataFrame
descriptive_stats_df = pd.concat(descriptive_stats_dict, axis=1)

# Display the descriptive statistics DataFrame
print(descriptive_stats_df)
```

```python
# Define the time periods for which to plot the data
time_periods = {
    '1925-2022': ('1925-01-01', '2022-12-31'),
    '1963-2022': ('1963-01-01', '2022-12-31'),
}

fig, axs = plt.subplots(len(time_periods), 2, figsize=(15, 10), sharex='col')

for i, (label, (start, end)) in enumerate(time_periods.items()):
    # Filter the data for the specific time period
    filtered_data = merged_data_final[(merged_data_final['date'] >= start) & (merged_data_final['date'] <= end)]
    
    # Group by date and compute mean and standard deviation of excess_return
    grouped_data = filtered_data.groupby('date')['excess_return'].agg(['mean', 'std']).reset_index()
    
    # Plot mean of excess_return
    axs[i, 0].plot(grouped_data['date'], grouped_data['mean'], label=f'Mean {label}')
    axs[i, 0].set_title(f'Mean of Excess Return: {label}')
    axs[i,0].set_xlabel('Date')
    axs[i, 0].set_ylabel('Mean')
    
    
    # Plot standard deviation of excess_return
    axs[i, 1].plot(grouped_data['date'], grouped_data['std'], label=f'Standard Deviation {label}', color='r')
    axs[i, 1].set_title(f'Standard Deviation of Excess Return: {label}')
    axs[i, 1].set_ylabel('Standard Deviation')
    axs[i,1].set_xlabel('Date')
    axs[i, 1].tick_params(axis='x', rotation=45)  # Rotate the x-axis labels
    
# Adjusting layout and setting common labels
plt.tight_layout()
plt.subplots_adjust(hspace=0.5)  # Increase the space between the plots to accommodate x-axis labels
plt.show()
```

## Compute monthly VWRETD descriptive stats and plot monthly

```python
# Define the time periods
time_periods = {
    '1925-2022': ('1925-01-01', '2022-12-31'),
    '1963-2022': ('1963-01-01', '2022-12-31'),
}

# Initialize a dictionary to store the descriptive stats for each time period
descriptive_stats_dict = {}

# Loop over each time period
for label, (start, end) in time_periods.items():
    # Filter the data for the specific time period
    filtered_data = merged_data_final[(merged_data_final['date'] >= start) & (merged_data_final['date'] <= end)]
    
    # Compute Excess Return and Log of Excess Return for MKT portfolio
    filtered_data['excess_return_mkt'] = filtered_data['vwretd'] - filtered_data['RF']
    filtered_data['log_excess_return_mkt'] = np.log(1 + filtered_data['excess_return_mkt'])
    
    # Compute and store descriptive stats
    descriptive_stats = filtered_data[['excess_return_mkt', 'log_excess_return_mkt']].describe(
        percentiles=[.01, .05, .25, .5, .75, .95, .99]
    )
    descriptive_stats.loc['skew'] = filtered_data[['excess_return_mkt', 'log_excess_return_mkt']].skew()
    descriptive_stats.loc['kurt'] = filtered_data[['excess_return_mkt', 'log_excess_return_mkt']].kurt()
    descriptive_stats_dict[label] = descriptive_stats

# Display the computed descriptive statistics
descriptive_stats_df = pd.concat(descriptive_stats_dict, axis=1)
print("Descriptive Statistics:\n", descriptive_stats_df)
```

## IPO return analysis

```python

# Reset the index
dsf_data.reset_index(inplace=True)

# Ensure the 'date' column is in datetime format.
dsf_data['date'] = pd.to_datetime(dsf_data['date'])

# Set 'PERMNO' and 'date' as indices again
dsf_data.set_index(['PERMNO', 'date'], inplace=True)

# Identify IPOs
ipos_data = dsf_data.groupby('PERMNO').first()

# Define time horizons in months.
time_horizons = {'1D': 0, '1M': 1, '12M': 12, '24M': 24, '36M': 36}

# Identify IPOs and their dates.
ipos_dates = dsf_data.reset_index().groupby('PERMNO')['date'].min()
```

```python
import concurrent.futures


# Define the function to process permnos
def process_permnos(permno, ipo_date, time_horizons, dsf_data):
    result = {}
    for label, months in time_horizons.items():
        end_date = ipo_date + pd.DateOffset(months=months)
        subset = dsf_data.loc[(permno, slice(ipo_date, end_date)), 'RET']
        return_sum = (1 + subset).prod() - 1  # Calculate cumulative return.
        
        # For market return, if it's 1 day, use .loc accessor to get the value on the IPO date, else sum over the range.
        market_return = dsf_data.loc[(permno, ipo_date), 'vwretd'] if months == 0 else dsf_data.loc[(permno, slice(ipo_date, end_date)), 'vwretd'].sum()
        excess_return_sum = return_sum - market_return
        
        result[label] = {
            'return_sum': return_sum,
            'excess_return_sum': excess_return_sum
        }
    return result


# Use ThreadPoolExecutor to parallelize the calculations
with concurrent.futures.ThreadPoolExecutor() as executor:
    futures = [executor.submit(process_permnos, permno, ipo_date, time_horizons, dsf_data) for permno, ipo_date in ipos_dates.items()]
    results = [future.result() for future in concurrent.futures.as_completed(futures)]
```

```python
# Convert results to DataFrames for further processing
returns_df = pd.DataFrame({label: [res[label]['return_sum'] for res in results] for label in time_horizons.keys()})
excess_returns_df = pd.DataFrame({label: [res[label]['excess_return_sum'] for res in results] for label in time_horizons.keys()})

descriptive_stats_dict = {label: returns_df[label].describe(percentiles=[.01, .05, .25, .5, .75, .95, .99]) for label in time_horizons.keys()}
excess_descriptive_stats_dict = {label: excess_returns_df[label].describe(percentiles=[.01, .05, .25, .5, .75, .95, .99]) for label in time_horizons.keys()}

# Adding skewness and kurtosis to the descriptive stats
for label in time_horizons.keys():
    descriptive_stats_dict[label].loc['skew'] = returns_df[label].skew()
    descriptive_stats_dict[label].loc['kurt'] = returns_df[label].kurt()
    excess_descriptive_stats_dict[label].loc['skew'] = excess_returns_df[label].skew()
    excess_descriptive_stats_dict[label].loc['kurt'] = excess_returns_df[label].kurt()

# Concatenate descriptive stats and display.
descriptive_stats_df = pd.concat(descriptive_stats_dict, axis=1)
excess_descriptive_stats_df = pd.concat(excess_descriptive_stats_dict, axis=1)

print("Descriptive Statistics for Returns:\n", descriptive_stats_df)
print("\nDescriptive Statistics for Excess Returns:\n", excess_descriptive_stats_df)
```

### Observations from Descriptive Statistics:

-   Variability in Returns:

    -   There is substantial variability in returns at different time horizons. The standard deviation indicates significant dispersion around the mean for both returns and excess returns, suggesting a high level of risk and uncertainty in post-IPO performance.
-   Skewness and Kurtosis:

    -   The skewness and kurtosis are notably high, indicating the presence of extreme values or outliers. This is indicative of the potential for extreme positive returns but also implies significant risks.
-   Negative Excess Returns:

    -   The mean excess returns are negative for longer time horizons (12M, 24M, 36M), suggesting that, on average, IPOs underperform the market in the long run after the initial listing.
-   Short Term Performance:

    -   In the short term (1D, 1M), the mean returns and excess returns are positive, indicating that IPOs tend to perform well initially.
-   Decreasing Mean Excess Returns:

    -   The decreasing mean excess returns over time imply that the relative performance of IPOs compared to the market tends to decline over time, potentially due to the realization of true firm value as the market gains more information about the firm.

### Considerations for Benchmarking:

-   Using aggregate market returns (VWRETD) may not be the most appropriate benchmark to compute the excess returns of IPOs due to the unique risk and return characteristics of newly listed firms.
-   A better benchmark might be a portfolio or an index that is more representative of the characteristics of IPOs, such as an index of similarly sized firms or firms from the same industry.
-   Alternatively, considering factors such as firm size, book-to-market ratios, and momentum, which are known to explain stock returns, might help in creating a more tailored benchmark.


### Conclusion:

The descriptive statistics reveal significant variations and potential outliers in IPO returns, with generally positive short-term returns and negative excess returns in the long run. Selecting an appropriate benchmark is crucial for accurately assessing the performance of IPOs, and industry-specific, size-specific, or risk-adjusted benchmarks may offer more meaningful comparisons than aggregate market returns.
